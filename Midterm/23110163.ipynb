```python
import numpy as np
import scipy.stats
import pandas as pd
import matplotlib
import matplotlib.pyplot as pp
import seaborn as sns
from IPython import display
from ipywidgets import interact, widgets
```

```python
# Tạo một ma trận 2x2 có các phần tử là 1.
matrix = np.ones((2, 2))
print(matrix)

# Tạo một mảng rồi sao chép (deep copy) sang một mảng khác.

og_array = np.array([2,3,1,1,0,1,6,3])
copy_array = og_array.copy()
print(og_array)
print(copy_array)

# Tạo một ma trận 3x3x3 với các phần tử ngẫu nhiên kiểu số thực.
random_matrix = np.random.rand(3, 3, 3)
print(random_matrix)

# Tạo một mảng có giá trị từ 10 đến 49.
array_10_49 = np.arange(10, 50)
print(array_10_49)

# Tạo một mảng, hiển thị phần tử đầu tiên của nó.
arr = np.array([2,3,1,1,0,1,6,3])
arr1 = arr[0]
print(arr1)

# Tạo một mảng, hiển thị các phần tử ở những vị trí lẻ.
arr = np.array([2,3,1,1,0,1,6,3])
arr_odd = arr[1::2]
print(arr_odd)

# Tạo một ma trận cấp 2, hiển thị hàng cuối cùng của nó.

matrix_2d = np.array([[1,2],[3,4]])
last_row = matrix_2d[-1]
print(last_row)

# Tạo một mảng, hiển thị tất cả phần tử bằng 2 hay bằng 10.
arr = np.array([2,3,1,1,0,1,6,3,10,12, 2,10])
output = []
for i in arr:
    if i == 2 or i == 10:
        output.append(i)

print(output)

# Tạo một mảng, gán phần tử thứ 5 có giá trị là 1.
arr = np.array([2,3,1,1,0,1,6,3])
arr[4] = 1 
print(arr)

# Tạo một ma trận cấp 2, đổi hàng cuối cùng thành các giá trị 1.

matrix_2x2 = np.array([[1,2],[3,4]])
matrix_2x2[-1] = 1
print(matrix_2x2)
```
