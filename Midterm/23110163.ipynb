Lọc các thông tin từ dữ liệu covid-europe các trường sau: iso_code, location, date, new_cases, new_vaccinations, population.

Tách các thông tin bị trùng lắp thành các dataframe con: country(iso_code, location, popluation), cases(iso_code, date, new_cases), vaccinations(iso_code, date, new_vaccinations). Loại bỏ những dòng thừa hoặc không có dữ liệu (không có nghĩa).

Tính tỉ lệ tổng số cases/population, cases/vaccinations của từng quốc gia. Sắp xếp theo chiều giảm dần theo tỉ lệ cases/population.

Lưu ý: Không cung cấp mật khẩu truy cập bài Test và đường dẫn lấy dữ liệu.

```python
data = pd.read_csv('covid-europe.csv')
data 

# Lọc các thông tin từ dữ liệu covid-europe các trường sau: iso_code, location, date, new_cases, new_vaccinations, population.
filter_data = data[['iso_code', 'location', 'date', 'new_cases', 'new_vaccinations', 'population']]
print(filter_data)

# Tách các thông tin bị trùng lắp thành các dataframe con: country(iso_code, location, popluation), cases(iso_code, date, new_cases), vaccinations(iso_code, date, new_vaccinations). Loại bỏ những dòng thừa hoặc không có dữ liệu (không có nghĩa).
country = filter_data[['iso_code', 'location', 'population']].drop_duplicates().dropna()
cases = filter_data[['iso_code', 'date', 'new_cases']].drop_duplicates().dropna()
vaccinations = filter_data[['iso_code', 'date', 'new_vaccinations']].drop_duplicates().dropna()

print(country)
print(cases)
print(vaccinations)


# Tính tỉ lệ tổng số cases/population, cases/vaccinations của từng quốc gia. Sắp xếp theo chiều giảm dần theo tỉ lệ cases/population.
for country_code in country['iso_code']:
    sum_cases = cases[cases['iso_code'] == country_code]['new_cases'].sum()
    population = country[country['iso_code'] == country_code]['population'].values[0]
    sum_vac = vaccinations[vaccinations['iso_code'] == country_code]['new_vaccinations'].sum()
    if population == 0: # zero check (n/0 = inf)
        cases_population = 0
    else:
        cases_population = sum_cases / population
        
    if sum_vac == 0: # zero check (n/0 = inf)
        cases_vaccinations = 0
    else:
        cases_vaccinations = sum_cases / sum_vac
    
    country.loc[country['iso_code'] == country_code, 'cases/population'] = cases_population
    country.loc[country['iso_code'] == country_code, 'cases/vaccinations'] = cases_vaccinations
    
sorted_country_c_p = country.sort_values(by='cases/population', ascending=False)
sorted_country_c_p = sorted_country_c_p[['iso_code', 'location', 'population', 'cases/population']]
print(sorted_country_c_p)

sorted_country_c_v = country.sort_values(by='cases/vaccinations', ascending=False)
sorted_country_c_v = sorted_country_c_v[['iso_code', 'location', 'population', 'cases/vaccinations']]
print(sorted_country_c_v)
```
