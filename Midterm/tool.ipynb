### 1.Khởi tạo Mảng 

- `ones`: tạo mảng toàn số 1. (`np.ones((2,3))` tạo ma trận 2x3 toàn số 1)
- `random`: tạo mảng ngẫu nhiên. (`np.random.rand(2,3)` tạo ma trận 2x3 với các số ngẫu nhiên từ 0 đến 1)
- `linspace`: tạo mảng với các giá trị chia đều nhau trong một khoảng xác định. (`np.linspace(0, 1, 5)` tạo mảng gồm 5 giá trị từ 0 đến 1)
- `arange`: tạo mảng với các giá trị trong một khoảng xác định với bước nhảy cụ thể. (`np.arange(0, 10, 2)` tạo mảng gồm các số từ 0 đến 10 với bước nhảy là 2)
- `zeros`: tạo mảng toàn số 0. (`np.zeros((3,4))` tạo ma trận 3x4 toàn số 0)
- `eye`: tạo ma trận đơn vị (identity matrix). (`np.eye(3)` tạo ma trận 3x3 với đường chéo chính là 1, các phần tử khác là 0)
- `full`: tạo mảng với tất cả các phần tử có cùng một giá trị xác định. (`np.full((2,3), 7)` tạo ma trận 2x3 với tất cả các phần tử là số 7)

### Phần 1: Các dạng câu hỏi mở rộng tương tự

Ngoài việc tính tổng và tỷ lệ, đề bài có thể yêu cầu bạn làm việc với **Thời gian** hoặc **Điều kiện lọc** phức tạp hơn.

#### Dạng 1: Thống kê theo thời gian (Time-series Analysis)

*Câu hỏi:* Thay vì tính tổng toàn bộ, hãy tính **tổng số ca mắc mới theo từng tháng** của từng quốc gia.

```python
# Chuyển đổi sang datetime
data['date'] = pd.to_datetime(data['date'])

data['month'] = data['date'].dt.to_period('M')
# Gom nhóm theo quốc gia và tháng
```

### Phần 2: Khung giàn bài (Framework) để giải mọi bài toán Pandas

Để làm bài nhanh và chuẩn (tránh dùng vòng lặp `for` như trong bài mẫu của bạn vì nó xử lý chậm với dữ liệu lớn), hãy tuân thủ quy trình 5 bước sau:

#### Bước 1: Load & Filter (Đọc và Lọc cột cần thiết)

Luôn bắt đầu bằng việc chỉ giữ lại những gì cần dùng để giảm tải bộ nhớ.

```python
df = pd.read_csv('file.csv')
cols_to_use = ['iso_code', 'location', 'date', 'new_cases', 'population']
df = df[cols_to_use]

```

#### Bước 2: Data Cleaning (Làm sạch)

Xử lý dữ liệu rác, trùng lặp hoặc thiếu.

```python
# Xóa trùng lặp
df = df.drop_duplicates()

# Xử lý NaN (Thay thế bằng 0 hoặc xóa dòng)
df = df.fillna(0) 
# Hoặc: df = df.dropna(subset=['iso_code'])

```

#### Bước 3: Group & Aggregate (Gom nhóm và Tính toán) **(Quan trọng nhất)**

Thay vì dùng vòng lặp `for` để chạy qua từng quốc gia (rất chậm), hãy dùng `groupby`. Đây là kỹ thuật "chia để trị" của Pandas.

* Muốn tính tổng? Dùng `.sum()`
* Muốn tính trung bình? Dùng `.mean()`
* Muốn đếm? Dùng `.count()`

```python
# Gom theo quốc gia, tính tổng ca mắc và tổng vắc xin cùng lúc
# Quy tắc: df.groupby('Cột muốn gom')['Cột muốn tính'].phép_tính()
stats = df.groupby(['iso_code', 'location', 'population'])[['new_cases', 'new_vaccinations']].sum().reset_index()

```

#### Bước 4: Feature Engineering (Tạo cột mới từ dữ liệu đã gom)

Thực hiện các phép chia, nhân, cộng trừ trên các cột. Pandas cho phép tính toán trên toàn bộ cột (vectorization) mà không cần loop.

```python
# Tính tỷ lệ (Tự động xử lý cho mọi dòng cùng lúc)
# Lưu ý: Nên xử lý chia cho 0 nếu cần
stats['cases_per_pop'] = stats['new_cases'] / stats['population']

```

#### Bước 5: Sort & Format (Sắp xếp và Trình bày)

```python
# Sắp xếp giảm dần
result = stats.sort_values(by='cases_per_pop', ascending=False)

# Chọn lại cột để hiển thị cuối cùng
final_view = result[['location', 'cases_per_pop']]
print(final_view)

```

### Tối ưu hóa code của bạn (Lời khuyên)

Trong đoạn code bạn gửi, đoạn dùng `for country_code in country['iso_code']:` là cách tư duy lập trình căn bản (C/Java). Trong Python/Pandas Data Science, người ta ưu tiên cách làm sau (gọn hơn và nhanh hơn gấp nhiều lần):

```python
# Cách làm chuẩn Pandas (Thay thế cho toàn bộ vòng lặp for của bạn)

# 1. Gom nhóm tính tổng cases và vaccinations theo iso_code
grouped = filter_data.groupby(['iso_code', 'population'])[['new_cases', 'new_vaccinations']].sum().reset_index()

# 2. Tính toán trực tiếp trên cột (Vectorization)
grouped['cases/population'] = grouped['new_cases'] / grouped['population']

# Xử lý chia cho 0 cho cột vaccinations (nếu sum_vac = 0)
# np.where(điều_kiện, giá_trị_nếu_đúng, giá_trị_nếu_sai)
grouped['cases/vaccinations'] = np.where(grouped['new_vaccinations'] == 0, 
                                         0, 
                                         grouped['new_cases'] / grouped['new_vaccinations'])

# 3. Sắp xếp
sorted_df = grouped.sort_values(by='cases/population', ascending=False)
print(sorted_df)

```

````python
data = pd.read_csv('covid-europe.csv')
data 

# Lọc các thông tin từ dữ liệu covid-europe các trường sau: iso_code, location, date, new_cases, new_vaccinations, population.
filter_data = data[['iso_code', 'location', 'date', 'new_cases', 'new_vaccinations', 'population']]
print(filter_data)

# Tách các thông tin bị trùng lắp thành các dataframe con: country(iso_code, location, popluation), cases(iso_code, date, new_cases), vaccinations(iso_code, date, new_vaccinations). Loại bỏ những dòng thừa hoặc không có dữ liệu (không có nghĩa).
country = filter_data[['iso_code', 'location', 'population']].drop_duplicates().dropna()
cases = filter_data[['iso_code', 'date', 'new_cases']].drop_duplicates().dropna()
vaccinations = filter_data[['iso_code', 'date', 'new_vaccinations']].drop_duplicates().dropna()

print(country)
print(cases)
print(vaccinations)


# Tính tỉ lệ tổng số cases/population, cases/vaccinations của từng quốc gia. Sắp xếp theo chiều giảm dần theo tỉ lệ cases/population.
for country_code in country['iso_code']:
    sum_cases = cases[cases['iso_code'] == country_code]['new_cases'].sum()
    population = country[country['iso_code'] == country_code]['population'].values[0]
    sum_vac = vaccinations[vaccinations['iso_code'] == country_code]['new_vaccinations'].sum()
    if population == 0: # zero check (n/0 = inf)
        cases_population = 0
    else:
        cases_population = sum_cases / population
        
    if sum_vac == 0: # zero check (n/0 = inf)
        cases_vaccinations = 0
    else:
        cases_vaccinations = sum_cases / sum_vac
    
    country.loc[country['iso_code'] == country_code, 'cases/population'] = cases_population
    country.loc[country['iso_code'] == country_code, 'cases/vaccinations'] = cases_vaccinations
    
sorted_country_c_p = country.sort_values(by='cases/population', ascending=False)
sorted_country_c_p = sorted_country_c_p[['iso_code', 'location', 'population', 'cases/population']]
print(sorted_country_c_p)

sorted_country_c_v = country.sort_values(by='cases/vaccinations', ascending=False)
sorted_country_c_v = sorted_country_c_v[['iso_code', 'location', 'population', 'cases/vaccinations']]
print(sorted_country_c_v)
````
