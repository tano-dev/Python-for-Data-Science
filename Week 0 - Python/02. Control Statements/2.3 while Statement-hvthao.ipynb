## 2.3 while Statement

The <b>while</b> statement allows you to repeat one or more actions while a condition remains True. Such a statement often is called a <b>loop</b>.

Let’s use a while statement to find the first power of 3 larger than 50:

```python
product = 2
```

```python
while product <= 2048:
    product = product * 2
```

```python
product
```

Something in the while statement’s suite must change product’s value, so the condition eventually becomes False. Otherwise, a logic error called an <b>infinite loop</b> occur.
