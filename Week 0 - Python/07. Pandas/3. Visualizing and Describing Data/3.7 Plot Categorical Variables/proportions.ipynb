```python
import numpy as np
import scipy.stats
import pandas as pd
```

```python
import matplotlib
import matplotlib.pyplot as pp

import pandas.plotting

from IPython import display
from ipywidgets import interact, widgets

%matplotlib inline
```

```python
import re
import mailbox
import csv
```

```python
smoking = pd.read_csv('whickham.csv')
```

```python
smoking.head()
```

```python
smoking['ageGroup'] = pd.cut(smoking.age,[0,30,40,53,64],labels=['0-30','30-40','40-53','53-64'])
```

```python
bysmoker = smoking.groupby("smoker").outcome.value_counts(normalize=True)
```

```python
byage = smoking.groupby(['ageGroup','smoker']).outcome.value_counts(normalize=True)
```

```python
bysmoker
```

```python
byage
```

```python
smoking.outcome.value_counts(normalize=True).plot(kind='bar', color=['C0','C3'])
```

```python
pp.figure(figsize=(10,3))
pp.subplot(1,2,1); pp.title('outcome'); smoking.outcome.value_counts().plot(kind='bar', color=['C0','C1'])
pp.subplot(1,2,2); pp.title('smoker'); smoking.smoker.value_counts().plot(kind='bar', color=['C2','C3'])
```

```python
pp.figure(figsize=(10,4))
pp.subplot(1,2,1); pp.title('outcome'); smoking.outcome.value_counts().plot(kind='barh', color=['C0','C1'])
pp.subplot(1,2,2); pp.title('smoker'); smoking.smoker.value_counts().plot(kind='barh', color=['C0','C1'])
```

```python
pp.figure(figsize=(10,4))
pp.subplot(1,2,1); smoking.outcome.value_counts().plot(kind='pie'); pp.title('outcome')
pp.subplot(1,2,2); smoking.smoker.value_counts().plot(kind='pie'); pp.title('smoker')
```
