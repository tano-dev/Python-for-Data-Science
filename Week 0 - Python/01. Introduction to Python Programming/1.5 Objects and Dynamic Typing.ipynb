## 1.5 Objects and Dynamic Typing

Values such as 7 (an integer), 4.1 (a floating-point number) and 'dog' are all objects. Every object has a type and a value:

```python
type(7)
```

```python
type(4.1)
```

```python
type('dog')
```

#### Variables Refer to Objects

Assigning an object to a variable binds (associates) that variable’s name to the object. As you’ve seen, you can then use the variable in your code to access the object’s value:

```python
x = 7
```

```python
x + 10
```

```python
x
```

After snippet [4]’s assignment, the variable x <b>refers</b> to the integer object containing 7. As shown in snippet [6], snippet [5] does not change x’s value. You can change x as follows: 

```python
x = x + 10
```

```python
x
```

#### Dynamic Typing

Python uses dynamic typing—it determines the type of the object a variable refers to while
executing your code. We can show this by rebinding the variable x to different objects and
checking their types:

```python
type(x)
```

```python
x = 4.1
```

```python
type(x)
```

```python
x = 'dog'
```

```python
type(x)
```

#### Garbage Collection

Python creates objects in memory and removes them from memory as necessary. After snippet [10], the variable x now refers to a float object. The integer object from snippet [7] is no longer bound to a variable. Python automatically removes such objects from memory. This process—called <b>garbage collection</b>—helps ensure that memory is available for new objects you create.
