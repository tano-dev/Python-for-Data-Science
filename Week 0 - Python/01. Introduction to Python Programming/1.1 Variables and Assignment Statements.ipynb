## 1.1 Variables and Assignment Statements

```python
 45 + 72
```

As in algebra, Python expressions also may contain <b>variables</b>, which store values for later
use in your code. Let’s create a variable named x that stores the integer 7, which is the variable’s <b>value</b>:

```python
x = 8
print(x)
```

```python
number1=int(input('Enter first integer:'))
```

```python
print(number1)
print(type(number1))
```

Snippet [2] is a <b>statement</b>. Each statement specifies a task to perform. The preceding statement creates x and uses the <b>assignment symbol</b> (=) to give x a value. The entire statement is an <b>assignment statement</b> that we read as “x is assigned the value 7.”

```python
y = 3
print(y)
```

#### Adding Variable Values and Viewing the Result

```python
x + y
```

#### Calculations in Assignment Statements

```python
total = x + y
```

```python
total
```

#### Variable Names

A variable name, such as x, is an <b>identifier</b>. Each identifier may consist of letters, digits and underscores (_) but may not begin with a digit. Python is <b>case sensitive</b>.
