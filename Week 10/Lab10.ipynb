1. Load the tb.csv file into a Panda's data frame and print out the column names.

```python
import pandas as pd
```

```python
data=pd.read_csv('tb.csv')
data
```

2. Apply the Panda's melt operaton. (new columns: sexage, cases)

```python
# 2. Apply the Panda's melt operaton. (new columns: sexage, cases)
data_melt=data.melt(data.columns[0:2],data.columns[2:,],'sexage','cases')
data_melt
```

3. Separate the combined sexage column into two (sex, age), using lambda.

```python
data_melt['sex'] = data_melt['sexage'].apply(lambda x: x[0])      # ký tự đầu
data_melt['age'] = data_melt['sexage'].apply(lambda x: x[1:])     # phần còn lại
data_melt
```

4. Rename the age range values, using Panda's map method. ('04'->'0-4', '514'-> '5-14', ...)

```python
# 4. Rename the age range values, using Panda's map method. ('04'->'0-4', '514'-> '5-14', ...)
age_map = {
    '04': '0-4',
    '514': '5-14',
    '1524': '15-24',
    '2534': '25-34',
    '3544': '35-44',
    '4554': '45-54',
    '5564': '55-64',
    '65plus': '65+',
    'u': 'unknown'
}
data_melt['age'] = data_melt['age'].map(age_map)
data_melt
```

5. Drop all the unknown values in the column cases, using dropna (with subset=[columns]).

```python
# 5. Drop all the unknown values in the column cases, using dropna (with subset=[columns]).
data_cleaned = data_melt.dropna(subset=['cases'])
data_cleaned
```

6. Sort the table by country, year, age, and sex using sort_values([columns]).

```python
# 6. Sort the table by country, year, age, and sex using sort_values([columns]).
data_sorted = data_cleaned.sort_values(['country', 'year', 'age', 'sex'])
data_sorted 
```

```python
data_VN = data_sorted[data_sorted['country'] == 'VN']
data_VN
```
