Dưới đây là dataset hosts.csv (các host - khách sạn/nhà/phòng) của dịch vụ Airbnb (cho thuê phòng/nhà). Mỗi mẫu tin chứa các thông tin về vị trí (kinh, vĩ độ); số lượng nhà/phòng của host; số lượng phòng ngủ, giường ngủ tương ứng với nhà/phòng; loại nhà/phòng; và dịch vụ TV.

- Loại nhà/phòng có các kiểu: Entire home/apartment, Hotel room, Private Room.

- Loại dịch vụ TV thì có kích cỡ TV (24 inch, ..., 85 inch); loại TV (TV, HDTV); dịch vụ trên tivi như Amazon Prime Video, Netflix, ...

Các em có nhiệm vụ làm sạch và biến đổi dữ liệu để tránh dư thừa dữ liệu: dịch vụ TV.

Sau đó, dựa vào các bảng (dataframe) đã phân rã, hãy tìm các host (thông tin về vị trí, số lượng phòng, loại phòng, ..) có dịch vụ TV là Disney+.

```python
import pandas as pd
```

```python
hosts = pd.read_csv('hosts.csv')
```

```python
hosts.info()
```

```python
hosts.head()
```

```python
hosts.drop([hosts.columns[0]], axis=1, inplace=True)
```

```python
for col in hosts.columns:
    print(col)
```

```python
hosts_short = pd.read_csv('hosts-short.csv')
hosts_short
```

```python
tv_cols = [col for col in hosts.columns if 'TV' in col or 'HDTV' in col]
tv_cols
```

```python
bmelt = hosts.melt(
    id_vars=['id','latitude','longitude','accommodates','bedrooms','beds','price'],
    value_vars=tv_cols,
    var_name='tv_feature',
    value_name='has_feature'
)

bmelt
```

```python
bmelt = bmelt[bmelt['has_feature'] == 1]
print("Number of rows after filtering:", len(bmelt))
bmelt
```

```python
hosts_disney = bmelt[bmelt['tv_feature'].apply(lambda x: 'Disney' in x)]
print("Number of rows with Disney feature:", len(hosts_disney))
hosts_disney
```
